```python
import matplotlib.pyplot as plt
import cv2
```

```python
path = r'C:\Users\carl\Documents\research\2024\ML\images\1WTC.jpg'

image = cv2.imread(path)
edges = cv2.Canny(image, 100, 200, 3, L2gradient=True)
#image = cv2.imread("1WTC.jpg")
```

```python
#cv2.imshow('image', image)
#cv2.imshow('edges', edges)
#cv2.waitKey(0)

plt.figure()
plt.title('image')
plt.imsave('1WTC.png', edges, cmap='gray', format='png')
plt.imshow(edges, cmap='gray')
plt.show()
```
